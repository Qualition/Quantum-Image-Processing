```python
# opening the file in read mode
my_file = open("red channel.txt", "r")
  
# reading the file
data = my_file.read()
  
# replacing end splitting the text 
# when newline ('\n') is seen.
red_data = data.split("\n")
my_file.close()
```

```python
# opening the file in read mode
my_file = open("blue channel.txt", "r")
  
# reading the file
data = my_file.read()
  
# replacing end splitting the text 
# when newline ('\n') is seen.
blue_data = data.split("\n")
my_file.close()
```

```python
# opening the file in read mode
my_file = open("green channel.txt", "r")
  
# reading the file
data = my_file.read()
  
# replacing end splitting the text 
# when newline ('\n') is seen.
green_data = data.split("\n")
my_file.close()
```
